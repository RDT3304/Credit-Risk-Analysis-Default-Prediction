```python
from pycaret.classification import plot_model

def plot_all_model_plots(model, scale=1, save=True, fold=None, fit_kwargs=None, plot_kwargs=None, groups=None, verbose=True, display_format='png'):
    """
    Plot all available plots for a given trained classification model using PyCaret's plot_model function.

    Parameters:
    - model: scikit-learn compatible object
        The trained model object.
    - scale: float, default=1
        Scale factor for the plots.
    - save: bool, default=False
        Whether to save the plots as image files.
    - fold: int or None, optional
        The fold number for cross-validation plots. If None, the default fold (fold 0) is used.
    - fit_kwargs: dict or None, optional
        Additional keyword arguments for the fit function (if retraining is required).
    - plot_kwargs: dict or None, optional
        Additional keyword arguments for plot customization.
    - groups: str or None, optional
        Group variable for plotting in case of group-based plots.
    - verbose: bool, default=True
        Whether to display the plot names as they are generated.
    - display_format: str or None, optional
        Format for displaying the plots (e.g., 'png', 'pdf').

    Returns:
    - None
    """

    # List of available plots
    available_plots = [
        'pipeline',
        'auc',
        'threshold',
        'pr',
        'confusion_matrix',
        'error',
        'class_report',
        'boundary',
        'rfe',
        'learning',
        'manifold',
        'calibration',
        'vc',
        'dimension',
        'feature',
        'feature_all',
        'parameter',
        'lift',
        'gain',
        'tree',
        'ks'
    ]

    for plot_type in available_plots:
        try:
            plot_model(model, plot=plot_type, scale=scale, save=save, fold=fold, fit_kwargs=fit_kwargs,
                       plot_kwargs=plot_kwargs, groups=groups, verbose=verbose, display_format=display_format)
        except Exception as e:
            print(f"Error plotting {plot_type} plot: {str(e)}")

```
